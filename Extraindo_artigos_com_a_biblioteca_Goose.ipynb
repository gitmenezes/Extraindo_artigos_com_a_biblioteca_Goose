Extraindo artigos com a biblioteca "Goose".

#### Instalação da Biblioteca
> ##### pip install goose3


```python
from goose3 import Goose as go #Importação da biblioteca
url = 'https://pt.wikipedia.org/wiki/Pandemia_de_COVID-19' # url de pesquisa
def extract_artigo(url):
  """
  url -> link do artigo
  
  """
  importa_artigo = go()
  artigo = importa_artigo.extract(url)
  return artigo.cleaned_text # Tag padrão com o texto
```

```python
extract_artigo(url)
```
